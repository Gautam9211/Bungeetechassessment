```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Bungeetech/internship-test2/master/input/question-1/main.csv'
df = pd.read_csv(url,index_col=0)
df = pd.DataFrame(df)
print(df.head(5))
```

```python
type(df)
```

```python
df.head(10)
```

```python
population=list(df.Population)
print(population)
```

```python
df.drop(["Population","Total"],axis=1, inplace = True)
```

```python
df.head()
```

```python
 df_new=df.groupby((df.index//10)*10).sum()
```

```python
df_new["Population"]= ""
```

```python
df_new.head()
```

```python
ind = []
for index, item in enumerate(df.index):
    if item%10==0:
        ind.append(index-1)
    
```

```python
start=ind[1:]
last = len(df.index)
print(start,last)
```

```python
Population_values = []
for x in start:
    Population_values.append(population[x])
Population_values.append(population[last-1])
print(Population_values)
```

```python
index_no = df_new.columns.get_loc("Population")
for index, item in enumerate(Population_values):
      df_new.iloc[index,index_no]=  item
```
